```groovy
Map VARIANT_IMPACT_COLORS = ["deletion": '#ED8B00', "duplication": '#333F48', "frameshift": '#A2AAAD',
                                 "fusion"  : '#0085CA', "indel": '#9467bd', "insertion": '#43B02A', "missense": '#002D72',
                                 "none"    : '#7f7f7f', "nonsense": '#bcbd22', "splice site": '#17becf', "unknown": '#D0D3D4']

List VARIANT_IMPACT = ["deletion", "duplication","frameshift","fusion", "indel", "insertion", "missense", "none", "nonsense", "unknown"]

def availableColors = VARIANT_IMPACT.collect { it ->
    Map mp = [:]
    mp.put('color', VARIANT_IMPACT_COLORS.get(it))
    mp.put('value', it)
    mp
}

println (availableColors)

beakerx.labels = VARIANT_IMPACT.collect {it}
beakerx.labelsWithAvailableColors = ["availableColors": availableColors.collect {it}]


OutputCell.HIDDEN
```

```groovy
%%javascript
require.config({paths:{'plotly':'//cdn.plot.ly/plotly-latest.min'}});
require(['plotly'], function(Plotly) {
    window.Plotly = Plotly;
});

```

```groovy
%%javascript

var labels = []
var colors = []

for (var i = 0; i < beakerx.labelsWithAvailableColors.availableColors.length; i++) {
    labels.push(beakerx.labelsWithAvailableColors.availableColors[i].value)
    colors.push(beakerx.labelsWithAvailableColors.availableColors[i].color)
}



var labelsMain = ["deletion","duplication", "frameshift","fusion"]
var labels2 = [["deletion", 1], ["duplication", 5], ["frameshift",14],["fusion",0]]
var values2 = [1,5,14,0]

var values = [17, 4, 3, 72, 9, 6, 212, 3, 4, 5],
    totalNum = 335;

beakerx.displayHTML(this, '<div id="plotlyDiv"></div>');
   
    var totalNum = 0
    
    var totalNum = values.reduce(function(pv, cv) {return pv + cv}, 0)

    var roundedValues = []
    for (var i = 0; i< values.length; i++) {
        roundedValues.push(Math.round(values[i]/totalNum * 100 * 10**1) / 10**1 + '%')
    }


    var data = [{
        values: values,
        labels: labels,
        ids: labels,
        text: [],
        textinfo: 'text',
        type: 'pie',
        marker: {
            colors: colors
        },
        sort: false,
        hole: .5,
        hovertemplate: '<b>Label</b>: %{label}<br>' +
        '<b>Percentage</b>: %{percent:.1%}<br>' +
                        '<b>total number</b>: %{value}' +
                         "<extra></extra>"
    }];
    var layout = {
        font: {
            family: 'Avenir',
            size: 14
        },
        hovermode:'closest',
        annotations: [
            {
                font: {
                    size: 36
                },
                showarrow: false,
                text: totalNum,
                x: 0.5,
                y: 0.5
            }
        ],
        showlegend: true,
        legend: {
            itemclick: false,
            itemdoubleclick: false
        },
        height: 450,
        width: 550
    }



Plotly.newPlot('plotlyDiv', data, layout, {displayModeBar: false});

```
