```groovy
%%javascript

var labels = [],
    colors = [],
    protein_effect_labels = [],
    protein_effect_colors = [];

for (var i = 0; i < beakerx.impact_list.impact_list.length; i++) {
    labels.push(beakerx.impact_list.impact_list[i].value)
    colors.push(beakerx.impact_list.impact_list[i].color)
}

for (var i = 0; i < beakerx.protein_effect.protein_effect.length; i++) {
    protein_effect_labels.push(beakerx.protein_effect.protein_effect[i].value)
    protein_effect_colors.push(beakerx.protein_effect.protein_effect[i].color)
}


var values = [17, 4, 3, 72, 9, 6, 212, 3, 4, 5, 25],
    protein_effect_values = [62, 38, 13, 29, 3, 43, 147];

drawPieChart(colors, labels, values, "impactChart")
drawPieChart(protein_effect_colors, protein_effect_labels, protein_effect_values, "proteinEffectChart")


function drawPieChart(colors, labels, values, containerId) {
    //beakerx.displayHTML(this, '<div id=' + containerId + '></div>');
   
    var totalNum = values.reduce(function(pv, cv) {return pv + cv}, 0)

    var roundedValues = []
    for (var i = 0; i< values.length; i++) {
        roundedValues.push(Math.round(values[i]/totalNum * 100 * 10**1) / 10**1 + '%')
    }


    var data = [{
        values: values,
        labels: labels,
        ids: labels,
        text: [],
        textinfo: 'text',
        type: 'pie',
        marker: {
            colors: colors
        },
        sort: false,
        hole: .5,
        hovertemplate: '<b>Label</b>: %{label}<br>' +
        '<b>Percentage</b>: %{percent:.1%}<br>' +
                        '<b>total number</b>: %{value}' +
                         "<extra></extra>"
    }];
    var layout = {
        font: {
            family: 'Avenir',
            size: 14
        },
        hovermode:'closest',
        annotations: [
            {
                font: {
                    size: 36
                },
                showarrow: false,
                text: totalNum,
                x: 0.5,
                y: 0.5
            }
        ],
        showlegend: true,
        legend: {
            itemclick: false,
            itemdoubleclick: false
        },
        height: 650,
        width: 750
    }



Plotly.newPlot(containerId, data, layout, {displayModeBar: false});
}



```
